###Для $s$ по порядку жениха с вероятностью $\frac{s}{n}$ принимаем его, с $1 -\frac{s}{n}$ отвергаем

```python
n = 10000

for j in range(10):
  taken_best = 0
  for tries in range(1000):
    candidates = np.array([rd.randint(1, n) for i in range(n)])
    best = max(candidates)
    for s in range(n):
      if rd.randint(1, n) <= s and candidates[s] == best:
        taken_best += 1
        break
  taken_best /= 1000
  print(taken_best)
```

###Выбираем случайного жениха

```python
n = 1000

for j in range(10):
  taken_best = 0
  for tries in range(1000):
    candidates = np.array([rd.randint(1, n) for i in range(n)])
    best = max(candidates)
    if candidates[rd.randint(0, n - 1)] == best:
      taken_best += 1
  taken_best /= 1000
  print(taken_best)
```
